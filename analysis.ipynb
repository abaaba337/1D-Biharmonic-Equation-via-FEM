```python
from meshpy.triangle import MeshInfo, build
import matplotlib.pyplot as plt
import matplotlib.patches as patches
import numpy as np
import sympy as sp
```

```python
# Draw Basis
# 创建一个数值范围
x = np.linspace(0, 1, 400)
x2 = np.linspace(1, 2, 400)

# 定义函数
def f1(x):
    return 2*x**3 - 3*x**2 + 1

def f2(x):
    return -2*x**3 + 3*x**2

def f3(x):
    return x**3 - 2*x**2 + x

def f4(x):
    return x**3 - x**2

def f5(x):
    x=x-1
    return 2*x**3 - 3*x**2 + 1

def f6(x):
    x=x-1
    return -2*x**3 + 3*x**2

def f7(x):
    x=x-1
    return x**3 - 2*x**2 + x

def f8(x):
    x=x-1
    return x**3 - x**2

# 计算函数值
y1 = f1(x)
y2 = f2(x)
y3 = f3(x)
y4 = f4(x)

y5 = f5(x2)
y6 = f6(x2)
y7 = f7(x2)
y8 = f8(x2)
```

```python
# 绘制图像
plt.figure(figsize=(10, 6))
plt.plot(x, y1, label='2x^3 - 3x^2 + 1', linewidth=3, color='#c1121f')  # 加粗线条
plt.plot(x, y2, label='-2x^3 + 3x^2', linewidth=3, color='#003049')    # 加粗线条
plt.plot(x, y3, label='x^3 - 2x^2 + x', linewidth=3, color='#386641')  # 加粗线条
plt.plot(x, y4, label='x^3 - x^2', linewidth=3, color='#5e548e')       # 加粗线条
plt.axhline(0, linestyle='--', linewidth=3, color='#949494')

# 添加标题和轴标签
plt.title('Graphs of Functions')
plt.xlabel('x')
plt.ylabel('y')

# 显示图像
plt.show()
```

```python
# Calculte local stiffness matrix
import sympy as sp
x, h = sp.symbols('x h')
D2phi = sp.Matrix([12*x - 6, 6*x - 4, -12*x + 6, 6*x - 2])

V = sp.diag(1, h, 1, h)
basis_product = D2phi * D2phi .transpose()
integral_matrix = basis_product.applyfunc(lambda elem: sp.integrate(elem, x))

integral_at_1 = integral_matrix.subs(x, 1)
integral_at_0 = integral_matrix.subs(x, 0)
result = V.T * (integral_at_1 - integral_at_0) * V 
```
